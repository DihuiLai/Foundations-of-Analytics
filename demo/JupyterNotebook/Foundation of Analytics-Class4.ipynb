## Linear Regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
import statsmodels.api as sm
```

```python
X=2*np.random.randn(1, 5000)
Y=.3*X+10+np.random.randn(1, 5000)
```

```python
X
```

```python
Y
```

```python
X_pd=pd.DataFrame(X.T)
Y_pd=pd.DataFrame(Y.T)
```

```python
X_pd
```

```python
X_pd = sm.add_constant(X_pd)
```

```python
X_pd
```

```python
# Instantiate a gaussian family model with the default link function.
gaussian_model = sm.GLM(Y_pd, X_pd, family=sm.families.Gaussian())

gaussian_model_results = gaussian_model.fit()
```

```python
print(gaussian_model_results.summary())
```

```python
reg = LinearRegression().fit(X.T, Y.T)
print([reg.coef_, reg.intercept_])
```
