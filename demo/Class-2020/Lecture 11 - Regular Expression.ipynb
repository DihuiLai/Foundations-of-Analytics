
## Regular Epxression

#### References:
https://www.w3schools.com/python/python_regex.asp

```python
import re 
str = "ckkd,,calal881 cc dkdk 01 ..alc1"
```

#### Find numeric values

```python
re.findall('[0-9]', str)
```

#### Find numeric value sets

```python
re.findall('[0-9]+', str) 
```

#### Find lower case letters

```python
re.findall('[a-z]', str) 
```

#### Find lower case letter sets

```python
re.findall('[a-z]+', str) 
```

#### Find number and letters

```python
re.findall('[a-zA-Z0-9]+', str)
```

#### Find word Harry in the novel Harry Potter

```python
file = open("data/J. K. Rowling - Harry Potter 1 - Sorcerer's Stone",'r')
raw_data = file.read()
file.close()
```

```python
len(re.findall('Harry', raw_data))
```

#### Find license plate number

```python
re.findall('[A-Za-z]{1,3}-[A-Za-z]{1,2}-[0-9]{1,4}', "ckckkkdkdkdkdkdk CIG-HC-1245 dccaaacc")
```
